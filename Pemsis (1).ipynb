```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

tahun = np.array([2019, 2020, 2021, 2022, 2023])
produksi = np.array([1374, 1508, 1386, 1544, 1506])

x = tahun - tahun.mean()
y = produksi


m = np.sum(x * y) / np.sum(x**2)
c = y.mean() - m * tahun.mean()


tahun_forecast = 2024
produksi_forecast = m * tahun_forecast + c


print("Hasil Trend Moment:")
print(f"Persamaan Regresi: Y = {m:.2f}X + {c:.2f}")
print(f"Prediksi Produksi Tahun {tahun_forecast}: {produksi_forecast:.2f} ton")
print(m)

#scatter
plt.figure(figsize=(10, 6))
plt.scatter(tahun, produksi, color='blue', label='Data Aktual')
plt.plot(tahun, m * tahun + c, color='red', label='Garis Tren (Regresi)')
plt.scatter(tahun_forecast, produksi_forecast, color='green', label=f'Prediksi {tahun_forecast}')
plt.title("Trend Produksi Cabai dan Prediksi Tahun 2024")
plt.xlabel("Tahun")
plt.ylabel("Produksi (ton)")
plt.legend()
plt.grid()
plt.show()
```
